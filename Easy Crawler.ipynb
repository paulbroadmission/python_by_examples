## sample code 你的第一支新聞爬蟲 ##

```python
import requests
from bs4 import BeautifulSoup

def get_hotline_udnnews(url):
    resp = requests.get(url)
    
    if resp.status_code is 200:
        resp.encoding = 'utf-8'
        soup = BeautifulSoup(resp.text, 'html.parser')
        scope = soup.select(".tab-link") #選取第一區段
        
        hot_lines = []
        
        for line in scope: #集成於一個list，後可回傳
          content = line.text.replace("\n","") #先取出全部 存在content 裡
          if ':' in content:  #如果有冒號就代表有時間在前(即時), UDN標題內文的冒號會是全形：所以誤判機率較小
            hot_lines.append(content)  #有時間的是即時, 後面編號的是熱門
            
        return hot_lines


if __name__ == '__main__':
    udnnews_lines = get_hotline_udnnews("https://udn.com/news/index")
    i = 0
    
    while i < len(udnnews_lines): 
        print(udnnews_lines[i])
        i += 1
```

## 分解資料方便儲存 ##
把時間與內容分離儲存

```python
import requests
from bs4 import BeautifulSoup

def get_hotline_udnnews(url):
    resp = requests.get(url)
    
    if resp.status_code is 200:
        resp.encoding = 'utf-8'
        soup = BeautifulSoup(resp.text, 'html.parser')
        soup = BeautifulSoup(resp.text, 'html.parser')
        scope = soup.select(".tab-link") #選取第一區段
        
        hot_lines = []
        
        for line in scope: #集成於一個list，後可回傳
          content = line.text.replace("\n","") #先取出全部 存在content 裡
          if ':' in content:  #如果有冒號就代表有時間在前(即時), UDN標題內文的冒號會是全形：所以誤判機率較小
            hot_lines.append(content)  #有時間的是即時, 後面編號的是熱門
            
        return hot_lines


if __name__ == '__main__':
    udnnews_lines = get_hotline_udnnews("https://udn.com/news/index")
    i = 0
    udnnews_dicts = {}
    while i < len(udnnews_lines):  #運用 dictionary達成分項的效果
        udnnews_dicts['time'] = udnnews_lines[i][:5]
        udnnews_dicts['title'] = udnnews_lines[i][5:]
        print("Time[{}]: {}".format(udnnews_dicts['time'], udnnews_dicts['title']))
        i += 1
```
